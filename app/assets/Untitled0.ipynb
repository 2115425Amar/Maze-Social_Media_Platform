```python
# prompt: generate 100000 random int between 100 to 200 which mimic stock behaviour in normal market   bear:bull    7:3  also display chart

import random
import matplotlib.pyplot as plt

def generate_stock_data(num_points, bear_bull_ratio):
    data = []
    current_price = 300  # Start at a reasonable price
    for _ in range(num_points):
        # Determine if the market moves up or down based on the bear:bull ratio
        if random.random() < bear_bull_ratio:
            # Bear market (downward trend)
            change = random.uniform(-2, 0)  # Smaller downward movements
        else:
            # Bull market (upward trend)
            change = random.uniform(0, 3)   # Larger upward movements

        # Apply the price change
        current_price += change
        # Ensure the price stays within a realistic range
        current_price = max(100, min(current_price, 200))

        data.append(current_price)
    return data

# Generate the stock data
num_points = 100000
bear_bull_ratio = 0.8
stock_data = generate_stock_data(num_points, bear_bull_ratio)

# Display the chart
plt.figure(figsize=(12, 6))
plt.plot(stock_data)
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.title('Simulated Stock Price with Bear/Bull Market')
plt.grid(True)
plt.show()
```
